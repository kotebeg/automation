# ChatGPT

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time



url = "https://chatgpt.com/?model=auto"
opts = Options()

# "C:\Program Files\Google\Chrome\Application\chrome.exe" --remote-debugging-port=9222 --user-data-dir="C:\chrome_debugging_profile"
opts.add_experimental_option("debuggerAddress", "localhost:9222")

driver = webdriver.Chrome(options = opts)

driver.get(url)
```

```python
def send_text_to_chatGPT_and_read_last_response(text):
    # Find the div with the id 'prompt-textarea'
    textarea_div = driver.find_element(By.ID, "prompt-textarea")

    # Execute JavaScript to add a <p> element with text inside the div
    driver.execute_script(f"""
        var div = arguments[0];
        var p = document.createElement('p');
        p.textContent = '{text}';  // Replace with the text you want
        div.appendChild(p);
    """, textarea_div)

    time.sleep(2)

    # Find the button with aria-label "Send prompt"
    send_button = driver.find_element(By.XPATH, '//*[@aria-label="Send prompt"]')

    # Click the Send button
    send_button.click()


    time.sleep(4)


    # Find all div elements with the class 'markdown prose w-full break-words dark:prose-invert dark'
    answer_divs = driver.find_elements(By.CSS_SELECTOR, 'div.markdown.prose.w-full.break-words.dark\\:prose-invert.dark')

    # Check if any elements were found
    if answer_divs:
        # Get the last element from the list
        last_answer_div = answer_divs[-1]
        
        # Get the text content of the last div
        last_answer_text = last_answer_div.text

        # Print the text of the last answer
        print(f"chatGPT Answer:<< {last_answer_text}")
    else:
        print("No answers found.")

    return last_answer_text
```

```python
chatGPT_responses = []
chatGPT_quiestions = [
    "What is capital of England",
    "What is capital of France",
    "What is capital of German",
    "What is capital of USA",
    "What is capital of Japan",
]
for i in chatGPT_quiestions: 
    time.sleep(1)
    print("question Send: >>", i)
    result = send_text_to_chatGPT_and_read_last_response(i)
    chatGPT_responses.append(result)
```
